```python
import pandas as pd
import numpy as np
import os

os.system("clear")
```

```python
ENDERECO_DADOS = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'
```

```python
try: 
    print('Obtendo os dados...')
    df_roubos = pd.read_csv(ENDERECO_DADOS, sep=';', encoding='iso-8859-1')    
    # display(df_roubos.head())
    # display(df_roubos['regiao'].unique())
    # display(df_roubos.info())
    # display(df_roubos.describe())  
    # display(df_roubos.columns)

except Exception as e:
    print(f'Erro ao obter os dados: {e}')
```

```python
try:    
    # delimitando as variaveis
    df_roubos = df_roubos[['cisp', 'regiao', 'munic', 'roubo_veiculo']]

    # correção dos dados originais
    df_correcao_acentuacao = df_roubos['regiao'].str.startswith('Grande Niter', na=False)
    # display(df_correcao_acentuacao)

    # Grava Grande Niteroi na posição onde é True
    df_roubos.loc[df_correcao_acentuacao, 'regiao'] = 'Grande Niterói'
    # display(df_roubos['regiao'].unique())

    # agrupando por região 'cisp', 'regiao' e 'munic'
    df_roubos = df_roubos.groupby(['cisp', 'regiao', 'munic'], as_index=False)['roubo_veiculo'].sum()
    # display(df_roubos.head())

except Exception as e:
    print(f'Erro ao limpar os dados: {e}')   
```

```python
# processando as informações
try:
    array_roubos = np.array(df_roubos['roubo_veiculo'])
    media =np.mean(array_roubos)
    mediana = np.median(array_roubos)
    total = np.sum(array_roubos)

    # Obter os quartis
    q1 = np.quantile(array_roubos, 0.25)
    q3 = np.quantile(array_roubos, 0.75) 

    display(f'Média: {media:.2f}')
    display(f'Mediana: {mediana:.2f}')
    display(f'Total: {total}')
    display(f'Q1: {q1}')
    display(f'Q3: {q3}')    



except Exception as e:
    print(f'Erro ao processar os dados: {e}')   
```

```python
    # Criando a coluna Flag (mais e menos)
try:
    # Maiores
    df_maiores = df_roubos[df_roubos['roubo_veiculo'] > q3].copy() # para não afetar o df_roubos original
    df_maiores['flag'] = 'Maiores'

    # Menores
    df_menores = df_roubos[df_roubos['roubo_veiculo'] < q1].copy() # para não afetar o df_roubos original
    df_menores['flag'] = 'Menores'  

    # Juntar os dataframs Maiores e Menores (concatenar)
    df_roubos_flag = pd.concat([df_maiores, df_menores], ignore_index=True) # interessante quando for exportar para outro programa 

    display(df_roubos_flag.head())    

except Exception as e:
    print(f'Erro ao criar a coluna Flag: {e}')
```

```python
    # Exportando os dados para CSV e/ou Excel
try:
    df_roubos_flag.to_csv('roubos_veiculos.csv', index=False)
    df_roubos_flag.to_excel('roubos_veiculos.xlsx', index=False)

    display('Dados exportados com sucesso!')    

except Exception as e:
    print(f'Erro ao exportar os dados: {e}')
```
